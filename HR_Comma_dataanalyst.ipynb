```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load dataset
df = pd.read_csv('HR_comma_sep.csv')

# Check data
df.head()
```

```python
cols = ['satisfaction_level','last_evaluation','number_project',
        'average_montly_hours','time_spend_company']

summary = pd.DataFrame()

summary['Mean'] = df[cols].mean()
summary['Median'] = df[cols].median()
summary['Mode'] = df[cols].mode().iloc[0]

summary['Range'] = df[cols].max() - df[cols].min()
summary['Variance'] = df[cols].var()
summary['Std Dev'] = df[cols].std()
summary['IQR'] = df[cols].quantile(0.75) - df[cols].quantile(0.25)

summary
```

```python
plt.hist(df['satisfaction_level'], bins=30)
plt.title("Histogram of Satisfaction Level")
plt.xlabel("Satisfaction Level")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x=df['satisfaction_level'])
plt.title("Box Plot of Satisfaction Level")
plt.show()
```

```python
stats.probplot(df['satisfaction_level'], dist="norm", plot=plt)
plt.title("Q-Q Plot")
plt.show()
```

```python
sns.pairplot(df, hue='left',
             vars=['satisfaction_level','last_evaluation',
                   'number_project','average_montly_hours',
                   'time_spend_company'])
plt.show()
```

```python
corr = df['last_evaluation'].corr(df['number_project'])
print("Correlation:", corr)
```
